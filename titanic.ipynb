```python
train = pd.read_csv('/kaggle/input/titanic/train.csv')
test = pd.read_csv('/kaggle/input/titanic/test.csv')
train.head()
```

```python
train.info()
test.info()
```

```python
train = train.drop('Cabin',axis=1)
test = test.drop('Cabin' ,axis=1)
```

```python
train.head()
```

```python
# Calculate the mean of the column
mean_value = train['Age'].mean()

# Replace missing values with the mean
train['Age'].fillna(mean_value, inplace=True)
train['Embarked'].fillna('Q', inplace=True)
```

```python
mean1 = test['Age'].mean()
median = test['Fare'].median()
test['Age'].fillna(mean1 , inplace=True)
test['Fare'].fillna(median , inplace=True)
```

```python
test.info()
```

```python
train.info()
```

```python
train = train.drop(['PassengerId','Name'], axis=1)
```

```python
test = test.drop(['PassengerId','Name','Ticket'], axis=1)
```

```python
train = train.drop(['Ticket'], axis=1)
```

```python
train['Sex'].value_counts()
```

```python
train['Ticket'].value_counts()
```

```python
train['Embarked'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# List of column names to apply label encoding
columns_to_encode = ['Sex', 'Embarked']

# Apply label encoding to each column
for column in columns_to_encode:
    train[column] = label_encoder.fit_transform(train[column])
    test[column] = label_encoder.fit_transform(test[column])

```

```python
X = train.drop('Survived',axis=1)
Y = train['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize the logistic regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Predict the target variable on the test data
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```

```python
Ypred = model.predict(test)
Ypred = np.array(Ypred)
```

```python
t2 = pd.read_csv('/kaggle/input/titanic/test.csv')
t2.head()
t2['Survived'] = Ypred 
```

```python
submission = pd.concat([t2['PassengerId'], t2['Survived']], axis=1)
```
